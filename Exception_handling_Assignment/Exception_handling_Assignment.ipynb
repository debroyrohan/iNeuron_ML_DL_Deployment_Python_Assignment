1. Write a function to compute 5/0 and use try/except to catch the exceptions.

```python
#Answer of question no. 1
def exception_compute(numerator, denominator):
    
    try:
        
        result = numerator / denominator
    
    except ZeroDivisionError:
        
        print("Denominator cannot be zero, its a zero division error")
        
    else:
        
        print(result)
        
    finally:
        
        print("Function executed")
        
#input
numerator = int(input("Please enter the numerator: "))
denominator = int(input("Please enter the denominator: "))

#function call
exception_compute(numerator, denominator)
```

2. Implement a Python program to generate all sentences where subject is in

["Americans", "Indians"] and verb is in ["Play", "watch"] and the object is in

["Baseball","cricket"].

Hint: Subject,Verb and Object should be declared in the program as shown below.

subjects=["Americans ","Indians"]

verbs=["play","watch"]

objects=["Baseball","Cricket"]

Output should come as below:

Americans play Baseball.

Americans play Cricket.

Americans watch Baseball.

Americans watch Cricket.

Indians play Baseball.

Indians play Cricket.

Indians watch Baseball.

Indians watch Cricket.


```python
#Answer of question no. 2

#This soltution I have designed based on OOPs methodology using abstaction and encapsulation concept.

#Subject Class
class Subject:
    
    def __init__(self, subject_list):
        self.__subject_list = subject_list
        
    def get_subject_list(self):
        return self.__subject_list
    
    def __str__(self):
        return "This is a Subject class"

#Verb Class
class Verb:
    
    def __init__(self, verb_list):
        self.__verb_list = verb_list
        
    def get_verb_list(self):
        return self.__verb_list
    
    def __str__(self):
        return "This is a Verb class"

#Object Class
class Object:
    
    def __init__(self, object_list):
        self.__object_list = object_list
        
    def get_object_list(self):
        return self.__object_list
    
    def __str__(self):
        return "This is a Object class"

#Sentence creation Class
class MakeSentence:
    
    def __init__(self, subject, verb, obj):
        self._subject = subject
        self._verb = verb
        self._obj = obj
        
    def get_subject(self):
        return self._subject
    
    def get_verb(self):
        return self._verb
    
    def get_obj(self):
        return self._obj
    
    def __str__(self):
        return "This class is used to make sentence"
    
    #Sentence generation method
    def generate_sentence(self):
        try:
            #To a sentence to be valid, it must have all these things i.e subject, verb & object
            if(len(self.get_subject().get_subject_list())
               and len(self.get_verb().get_verb_list())
               and len(self.get_obj().get_object_list())):
                
                for subject in self.get_subject().get_subject_list():
                    for verb in self.get_verb().get_verb_list():
                        for obj in self.get_obj().get_object_list():
                            print(subject.capitalize() + " " + verb.lower() + " " + obj.capitalize())
                
            else:
                return "All three components are required to make a sentence"
        except Exception as e:
            print("An Error has occured: {}".format(str(e)))
        
#inputs:
subject_list = list(map(str, input("Please enter the subject_list: ").split()))
verb_list = list(map(str, input("Please enter the verb_list: ").split()))
object_list = list(map(str, input("Please enter the object_list: ").split()))

subject = Subject(subject_list)
verb = Verb(verb_list)
obj = Object(object_list)

make_sentence = MakeSentence(subject, verb, obj)

make_sentence.generate_sentence()
```
